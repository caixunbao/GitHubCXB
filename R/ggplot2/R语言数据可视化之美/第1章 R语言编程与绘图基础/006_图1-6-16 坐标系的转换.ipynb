```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.pointsize=8,warn = -1)
library(ggplot2)
library(ggthemes)
```

```r
df <- read.csv("./data/PolarBar_Data.csv",header = TRUE)
```

```r
head(df)
```

```r
ggplot(data = df,mapping = aes(x = Date,y = Value))+
    geom_bar(stat = "identity",width = 10,color="black",size=0.25,fill="#3bc8e8")+
    scale_x_continuous(name = "Time(Date)",breaks = seq(0,360,30))+
    scale_y_continuous(breaks = seq(0,160,20),limits = c(0,160),expand = expand_scale(add = 0))+
    ggthemes::theme_economist()
```

```r
ggplot(data = df,mapping = aes(x = Date,y = Value))+
    geom_bar(stat = "identity", width = 10,colour="black",size=0.25,fill="#3BC8EB")+
    scale_x_continuous(breaks = seq(0,360,30))+
    scale_y_continuous(breaks = seq(0,160,20),limits = c(-5,115))+
    coord_polar(theta = "x",start = 0)+
    theme_light()+
    theme(panel.background = element_blank(),
          panel.grid.major = element_line(colour = "grey60",size = 0.25),
          axis.text.y = element_blank(),
          axis.ticks = element_blank())
```

```r
df <- read.csv("./data/PolarBar_Data.csv",header = T)
head(df)
```

```r
ggplot(data = df,mapping = aes(x = Date,y = Value))+
    geom_area(colour="black",size=0.25,fill="#FFA1B9")+
    scale_x_continuous(name = "Time(Date)",breaks = seq(from = 0,to = 360,by = 60),expand = expand_scale(add = 0))+
    scale_y_continuous(breaks = seq(0,120,20),limits = c(0,120),expand = expand_scale(add = 0))+
    theme_economist()
```

```r
ggplot(data = df,mapping = aes(Date,Value))+
    geom_area(color=NA,size=0.25,fill="#ffa1b9")+
    geom_line(color="black",size=0.25,group=1)+
    scale_x_continuous(name = "Time(Day)",breaks = seq(0,360,30),expand = expand_scale(add = 0))+
    scale_y_continuous(breaks = seq(0,120,20),limits = c(0,120),expand = expand_scale(add = 0))+
    coord_polar(theta = "x",start = 0)+
    theme_economist()+
    theme( panel.background = element_blank(),
         panel.grid.major = element_line(colour = "grey80",size=.25),
         axis.text.y = element_text(size = 10,colour="black"),
         axis.line.y = element_line(size=0.25),
         axis.text.x=element_text(size = 11,colour="black"))
```
