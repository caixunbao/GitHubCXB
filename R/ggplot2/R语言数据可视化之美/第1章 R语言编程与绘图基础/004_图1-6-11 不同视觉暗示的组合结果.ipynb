```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.pointsize=8,warn = -1)
library(ggplot2)
library(gridExtra)
df <- read.csv("./data/MappingAnalysis_Data.csv",header = TRUE)
```

```r
head(df)
```

```r
#--------------------------------------Size---------------------------
```

```r
ggplot(data = df,mapping = aes(x = Time,y = value,group=variable))+
    geom_line()+
    geom_point(shape=21,size=4,color="black",fill="white")+
    theme_classic()+
    theme(
    text=element_text(size=14,color="black"),
    plot.title=element_text(size=14,family="myfont",face="bold.italic",hjust=.5,color="black"),
    legend.background = element_blank(),
    legend.position=c(0.2,0.8))
```

```r
#--------------------------------------color---------------------------
```

```r
ggplot(data = df,mapping = aes(x = Time,y = value,group=variable,fill=variable))+
    geom_line()+
    geom_point(color="black",shape=21,size=4)+
    scale_fill_manual(values = c("grey60","grey30","black","white"))+
    theme_classic()+
    theme(
    text=element_text(size=14,color="black"),
    plot.title=element_text(size=14,family="myfont",face="bold.italic",hjust=.5,color="black"),
    legend.background = element_blank(),
    legend.position=c(0.2,0.8))
```

```r
#--------------------------------------shape---------------------------
```

```r
ggplot(data = df,mapping = aes(x = Time,y = value,shape=variable))+
    geom_line()+
    geom_point(size=4,color="black",fill="grey60")+
    scale_shape_manual(values = c(21,22,23,24))+
    theme_classic()+
    theme(
    text=element_text(size=14,color="black"),
    plot.title=element_text(size=14,family="myfont",face="bold.italic",hjust=.5,color="black"),
    legend.background = element_blank(),
    legend.position=c(0.2,0.8))
```
