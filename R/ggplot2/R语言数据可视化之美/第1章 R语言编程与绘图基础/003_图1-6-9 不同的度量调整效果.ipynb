```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.pointsize=8,warn = -1)
```

```r
library(ggplot2)
library(gridExtra)
library(RColorBrewer)
df <- read.csv("./data/Facet_Data.csv")
```

```r
p1<-ggplot(data = df,mapping = aes(x = SOD,y = tau,size=age))+
    geom_point(shape=21,color="black",fill="#e53f2f",stroke=0.25,alpha=0.8)+
    scale_size_continuous(range = c(0.5,8.5))
p1
```

```r
p2<-ggplot(data = df,mapping = aes(x = SOD,y = tau,size=age,fill=age))+
    geom_point(shape=21,color="black",stroke=0.25,alpha=0.8)+
    scale_size_continuous(range = c(1,8))+
    scale_fill_distiller(palette="Reds")
p2
```

```r
p3<-ggplot(df, aes(x=SOD,y=tau,fill=Class,shape=Class,size=Class)) + 
  geom_point(colour="black",stroke=0.25)+
  scale_fill_manual(values=c("#36BED9","#FF0000","#FBAD01"))+
  scale_shape_manual(values=c(21,22,23))
p3
```

```r
p4<-ggplot(data = df,mapping = aes(SOD,tau,fill=Class,size=age))+
    geom_point(colour="black",alpha=0.8,stroke=0.25,shape=21)+
    scale_fill_manual(values = c("#36BED9","#FF0000","#FBAD01"))+
    scale_size_continuous(range = c(1,8))
p4
```

```r
library(gridExtra)
```

```r
grid.arrange(p1,p2,p3,p4,ncol=2)
```
