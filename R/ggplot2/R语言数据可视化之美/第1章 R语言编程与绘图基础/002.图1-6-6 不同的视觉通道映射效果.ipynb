```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.pointsize=8,warn = -1)
library(ggplot2)
```

```r
df <- read.csv("./data/Facet_Data.csv")
```

```r
head(df)
```

```r
p1<-ggplot(data = df,mapping = aes(x = SOD,y = tau,size=age))+
    geom_point(shape=21,color="black",fill="#336a97",stroke=0.25)
p1
```

```r
p2<-ggplot(data = df,mapping = aes(x = SOD,y = tau,size=age))+
    geom_point(shape=21,color="black",fill="#336a97",stroke=0.25,alpha=0.8)
p2
```

```r
p3<-ggplot(df, aes(x=SOD,y=tau,fill=Class)) + 
  geom_point(shape=21,size=3,colour="black",stroke=0.25)
p3
```

```r
p4<-ggplot(df, aes(SOD,tau,fill=Class,size=age)) + 
  geom_point(shape=21,colour="black",stroke=0.25,
             alpha=0.8)
p4
```

```r
library(gridExtra)
```

```r
grid.arrange(p1,p2,p3,p4,ncol=2)
```
