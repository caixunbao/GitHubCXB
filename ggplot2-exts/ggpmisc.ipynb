```r
library(ggpmisc)
library(ggplot2);
```

```r
options(repr.plot.width=10, repr.plot.height=5, repr.plot.pointsize=8)
```

```r
formula <- y ~ x + I(x^2)
ggplot(data = cars,mapping = aes(x = speed,y = dist))+
    geom_point()+
    stat_fit_deviations(method = "lm",formula = formula,color="red")+
    geom_smooth(method = "lm",formula = formula)
```
